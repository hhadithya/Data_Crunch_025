```python
import pandas as pd
import numpy as np

def process_kingdom_group(train_df, kingdom_group_value):
    """
    Processes the dataset by averaging specified columns for a given kingdom_group value.
    It replaces values with their averaged counterparts and removes duplicate dates.

    Parameters:
        train_df (pd.DataFrame): The input dataframe containing the data.
        kingdom_group_value (int): The kingdom_group value to filter and process.

    Returns:
        pd.DataFrame: The modified dataframe with updated values.
    """
    # Define columns to average
    columns_to_average = [
        'Avg_Temperature',
        'Avg_Feels_Like_Temperature',
        'Temperature_Range',
        'Feels_Like_Temperature_Range',
        'Radiation',
        'Rain_Amount',
        'Rain_Duration',
        'Wind_Speed',
        'Wind_Direction',
        'Evapotranspiration',
    ]

    # Filter for the given kingdom_group value
    kg_df = train_df[train_df['kingdom_group'] == kingdom_group_value].copy()

    # Group by date and calculate means for specified columns
    grouped = kg_df.groupby('Date')[columns_to_average].transform('mean')

    # Update the values in the filtered dataframe
    kg_df[columns_to_average] = grouped

    # Get unique indices - the first occurrence of each date
    unique_indices = kg_df.drop_duplicates('Date').index

    # Get duplicate indices to drop
    duplicate_indices = kg_df.index.difference(unique_indices)

    # Update the values for the specified kingdom_group in the original dataframe
    train_df.loc[kg_df.index, columns_to_average] = kg_df[columns_to_average]

    # Drop the duplicate rows
    train_df = train_df.drop(duplicate_indices)

    return train_df
```

```python
import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize

# Define numerical columns (excluding non-relevant ones)
num_cols = ["Avg_Temperature", "Avg_Feels_Like_Temperature", "Temperature_Range",
            "Feels_Like_Temperature_Range", "Radiation", "Rain_Amount", "Rain_Duration",
            "Wind_Speed", "Evapotranspiration"]

def winsorize_outliers(df, num_cols, lower_percentile=1, upper_percentile=99):
    df_winsorized = df.copy()

    # Apply Winsorizing separately for each kingdom_group
    for kingdom in df["kingdom_group"].unique():
        mask = df["kingdom_group"] == kingdom

        for col in num_cols:
            lower_bound = np.percentile(df.loc[mask, col], lower_percentile)
            upper_bound = np.percentile(df.loc[mask, col], upper_percentile)

            df_winsorized.loc[mask, col] = np.clip(df.loc[mask, col], lower_bound, upper_bound)

            print(f"Applied Winsorizing on {col} for kingdom_group {kingdom}")

    return df_winsorized

# Apply the function to the dataset
train_df = winsorize_outliers(train_df, num_cols)
```

```python
import pandas as pd

# Function to find missing dates for each kingdom
def find_missing_dates(df):
    missing_dates_dict = {}

    for kingdom, dates in df.groupby('kingdom_group')['Date']:
        min_date = dates.min()
        max_date = dates.max()
        date_range = pd.date_range(min_date, max_date)

        actual_dates = set(dates)
        missing_dates = [date for date in date_range if date not in actual_dates]

        if missing_dates:
            missing_dates_dict[kingdom] = missing_dates

    return missing_dates_dict

# Find missing dates in training data
missing_dates_train = find_missing_dates(train_df)
print("Missing dates in training data:", missing_dates_train)

# Find missing dates in test data if it exists
if 'test_df' in locals():
    missing_dates_test = find_missing_dates(test_df)
    print("Missing dates in test data:", missing_dates_test)
```

```python
import pandas as pd
import numpy as np

# Assuming 'Year', 'Month', and 'Day' columns exist in your DataFrames
train_df['Date'] = pd.to_datetime(train_df[['Year', 'Month', 'Day']])
test_df['Date'] = pd.to_datetime(test_df[['Year', 'Month', 'Day']])

def is_leap_year(year):
    """Check if a given year is a leap year."""
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def cyclic_features(df):
    df['DayOfWeek_sin'] = np.sin(2 * np.pi * df['Date'].dt.dayofweek / 7)
    df['DayOfWeek_cos'] = np.cos(2 * np.pi * df['Date'].dt.dayofweek / 7)

    # Adjust for leap years dynamically
    df['DaysInYear'] = df['Date'].dt.year.apply(lambda y: 366 if is_leap_year(y) else 365)

    df['DayOfYear_sin'] = np.sin(2 * np.pi * df['Date'].dt.dayofyear / df['DaysInYear'])
    df['DayOfYear_cos'] = np.cos(2 * np.pi * df['Date'].dt.dayofyear / df['DaysInYear'])

    df['Month_sin'] = np.sin(2 * np.pi * df['Date'].dt.month / 12)
    df['Month_cos'] = np.cos(2 * np.pi * df['Date'].dt.month / 12)

    df['Quarter_sin'] = np.sin(2 * np.pi * df['Date'].dt.quarter / 4)
    df['Quarter_cos'] = np.cos(2 * np.pi * df['Date'].dt.quarter / 4)

    df.drop(columns=['DaysInYear'], inplace=True)  # Clean up
    return df

train_df = cyclic_features(train_df)
test_df = cyclic_features(test_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
import pandas as pd
import numpy as np

# To calculate metrics, you'll need a validation set
# Let's split the original training data
from sklearn.model_selection import train_test_split

# Assuming train_data is your full training dataset
train_subset, val_subset = train_test_split(train_data, test_size=0.001, random_state=None)

# Dictionary to store models and their performance metrics
models = {}
metrics = {}

# Target variables to predict
targets = ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed', 'Wind_Direction']

# Features to use for prediction
features = ['kingdom_group', 'DayOfWeek_sin','DayOfWeek_cos','DayOfYear_sin','DayOfYear_cos','Month_sin','Month_cos','Quarter_sin','Quarter_cos','Season']

# Train and evaluate a model for each target
for target in targets:
    print(f"Training and evaluating model for {target}")

    # Create and train the model on training subset
    model = RandomForestRegressor(n_estimators=100, random_state=None)
    model.fit(train_subset[features], train_subset[target])

    # Store the model
    models[target] = model

    test_data[target] = model.predict(test_data[features])

    # Make predictions on validation set
    val_predictions = model.predict(val_subset[features])

    # Calculate metrics
    mae = mean_absolute_error(val_subset[target], val_predictions)
    mse = mean_squared_error(val_subset[target], val_predictions)
    rmse = sqrt(mse)
    r2 = r2_score(val_subset[target], val_predictions)

    # Calculate MAPE (Mean Absolute Percentage Error)
    # Avoiding division by zero
    actual = val_subset[target].values
    mape = np.mean(np.abs((actual - val_predictions) / np.maximum(np.abs(actual), 1e-10))) * 100

    # Store metrics
    metrics[target] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R²': r2,
        'MAPE (%)': mape
    }

# Display metrics as a DataFrame
metrics_df = pd.DataFrame(metrics)
print("\nModel Performance Metrics:")
print(metrics_df)

# Average performance across all targets
print("\nAverage Performance Across All Targets:")
avg_metrics = metrics_df.mean(axis=1)
for metric, value in avg_metrics.items():
    print(f"{metric}: {value:.4f}")

scaled_predictions = test_data[targets].values
original_scale_predictions = scaler.inverse_transform(scaled_predictions)

for i, target in enumerate(targets):
    test_data[target] = original_scale_predictions[:, i]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import tensorflow as tf
import math
import time
from tqdm.notebook import tqdm

# Setting seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Target variable to predict
target_column = 'Avg_Temperature'

# Selected features based on previous analysis
selected_features = ['Rain_Amount']

# Function to add lag features
def add_lag_features(df, column, lags):
    df_copy = df.copy()
    for lag in lags:
        df_copy[f'{column}_lag_{lag}'] = df_copy[column].shift(lag)
    return df_copy

# Function to create sequences for LSTM
def create_sequences(X, y, time_steps=5):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

# Function to build and train LSTM model
def build_and_train_lstm_model(X_train_seq, y_train_seq, epochs=100, batch_size=32):
    model = Sequential()
    model.add(LSTM(64, activation='relu', return_sequences=True,
                   input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
    model.add(Dropout(0.2))
    model.add(LSTM(32, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    # Train the model
    history = model.fit(
        X_train_seq, y_train_seq,
        epochs=epochs,
        batch_size=batch_size,
        validation_split=0.2,
        verbose=0,
        shuffle=False,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=10,
                restore_best_weights=True
            )
        ]
    )

    return model, history

# Function to make predictions for future dates
def predict_future(model, last_sequence, scaler_X, scaler_y, future_dates, feature_values):
    """
    Make future predictions using a trained LSTM model

    Parameters:
    - model: Trained LSTM model
    - last_sequence: Last known sequence of data (used as starting point)
    - scaler_X: Fitted feature scaler
    - scaler_y: Fitted target scaler
    - future_dates: Array of dates to predict for
    - feature_values: Dictionary with feature values for future dates
                     (Keys are feature names, values are arrays)

    Returns:
    - DataFrame with dates and predictions
    """
    predictions = []
    sequence = last_sequence.copy()

    # For each future date
    for i, date in enumerate(future_dates):
        # Make prediction with current sequence
        pred_scaled = model.predict(sequence.reshape(1, sequence.shape[0], sequence.shape[1]), verbose=0)
        # Convert prediction back to original scale
        pred = scaler_y.inverse_transform(pred_scaled)[0][0]

        # Store prediction
        predictions.append(pred)

        # If we have more dates to predict, update the sequence
        if i < len(future_dates) - 1:
            # Create new data point with predicted value and features for next date
            new_point = np.zeros((1, sequence.shape[1]))

            # Set future feature values
            for j, feature in enumerate(selected_features):
                if feature in feature_values:
                    new_point[0, j] = feature_values[feature][i+1]
                else:
                    # If feature is not provided, use the value from the last sequence
                    new_point[0, j] = sequence[-1, j]

            # Update lag features with the prediction
            lag_features = [f'{target_column}_lag_{lag}' for lag in range(1, 6)]
            for j, lag_feature in enumerate(lag_features):
                if j == 0:
                    # The first lag is the prediction we just made
                    idx = selected_features.index(lag_feature) if lag_feature in selected_features else len(selected_features) + j
                    new_point[0, idx] = pred_scaled[0][0]
                else:
                    # Other lags are shifted from the sequence
                    idx = selected_features.index(lag_feature) if lag_feature in selected_features else len(selected_features) + j
                    prev_idx = selected_features.index(lag_features[j-1]) if lag_features[j-1] in selected_features else len(selected_features) + (j-1)
                    new_point[0, idx] = sequence[-1, prev_idx]

            # Scale the new point
            new_point_scaled = scaler_X.transform(new_point)

            # Update sequence by removing the first timestep and adding the new point
            sequence = np.vstack([sequence[1:], new_point_scaled])

    # Create DataFrame with results
    results_df = pd.DataFrame({
        'Date': future_dates,
        target_column: predictions
    })

    return results_df

# Main function to process all kingdom groups
def predict_all_kingdom_groups(train_df, test_df, lags_to_use=5):
    # Get unique kingdom groups
    kingdom_groups = train_df['kingdom_group'].unique()

    # Prepare DataFrame to store all predictions
    all_predictions = []

    # Process each kingdom group
    for kg in tqdm(kingdom_groups, desc="Processing kingdom groups"):
        # Filter data for this kingdom group
        kg_data = train_df[train_df['kingdom_group'] == kg].copy()
        kg_data['Date'] = pd.to_datetime(kg_data['Date'])
        kg_data = kg_data.sort_values('Date')

        # Get future dates for this kingdom group from test_df
        kg_test = test_df[test_df['kingdom_group'] == kg].copy()
        kg_test['Date'] = pd.to_datetime(kg_test['Date'])
        kg_test = kg_test.sort_values('Date')
        future_dates = kg_test['Date'].values

        # Skip if no test data for this kingdom group
        if len(future_dates) == 0:
            continue

        # Add lag features
        lags = list(range(1, lags_to_use + 1))
        kg_data_with_lags = add_lag_features(kg_data, target_column, lags)

        # Drop rows with NaN values
        kg_data_with_lags = kg_data_with_lags.dropna()

        # Create lag feature names
        lag_features = [f'{target_column}_lag_{lag}' for lag in lags]

        # Combined features
        final_features = selected_features + lag_features

        # Scale the data
        scaler_X = MinMaxScaler()
        scaler_y = MinMaxScaler()

        X = kg_data_with_lags[final_features].values
        y = kg_data_with_lags[target_column].values.reshape(-1, 1)

        X_scaled = scaler_X.fit_transform(X)
        y_scaled = scaler_y.fit_transform(y)

        # Create sequences for LSTM
        time_steps = 5
        X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)

        # Build and train model
        model, history = build_and_train_lstm_model(X_seq, y_seq)

        # Get last known sequence for predictions
        last_sequence = X_scaled[-time_steps:]

        # Create a dictionary with future feature values (where available)
        future_features = {}
        for feature in selected_features:
            if feature in kg_test.columns:
                future_features[feature] = kg_test[feature].values
            elif feature in ['DayOfYear_cos', 'DayOfYear_sin', 'Month_cos', 'Month_sin',
                            'Quarter_cos', 'Quarter_sin', 'DayOfWeek_cos', 'DayOfWeek_sin']:
                # These are cyclic features we can compute
                # Example for DayOfYear_cos:
                if feature == 'DayOfYear_cos':
                    day_of_year = kg_test['Date'].dt.dayofyear.values
                    future_features[feature] = np.cos(2 * np.pi * day_of_year / 365)
                elif feature == 'DayOfYear_sin':
                    day_of_year = kg_test['Date'].dt.dayofyear.values
                    future_features[feature] = np.sin(2 * np.pi * day_of_year / 365)
                elif feature == 'Month_cos':
                    month = kg_test['Date'].dt.month.values
                    future_features[feature] = np.cos(2 * np.pi * month / 12)
                elif feature == 'Month_sin':
                    month = kg_test['Date'].dt.month.values
                    future_features[feature] = np.sin(2 * np.pi * month / 12)
                # Add other cyclic features as needed

        # Make future predictions
        kg_predictions = predict_future(
            model, last_sequence, scaler_X, scaler_y,
            future_dates, future_features
        )

        # Add kingdom group to predictions
        kg_predictions['kingdom_group'] = kg

        # Append to all predictions
        all_predictions.append(kg_predictions)

        # Optionally plot predictions for this kingdom group
        plt.figure(figsize=(12, 5))
        plt.plot(kg_predictions['Date'], kg_predictions[target_column])
        plt.title(f'Predictions for Kingdom Group: {kg}')
        plt.xlabel('Date')
        plt.ylabel(target_column)
        plt.grid(True)
        plt.show()

    # Combine all predictions
    if all_predictions:
        combined_predictions = pd.concat(all_predictions, ignore_index=True)
        return combined_predictions
    else:
        return pd.DataFrame()

# Fill test_df with predictions
def fill_test_df_with_predictions(test_df, predictions_df):
    # Create a copy of test_df
    filled_test_df = test_df.copy()

    # Merge predictions with test_df
    filled_test_df = filled_test_df.merge(
        predictions_df[['Date', 'kingdom_group', target_column]],
        on=['Date', 'kingdom_group'],
        how='left'
    )

    return filled_test_df

# Main execution
# Run the prediction for all kingdom groups
predictions = predict_all_kingdom_groups(train_df, test_df)

# Fill test_df with predictions
filled_test_df = fill_test_df_with_predictions(test_df, predictions)

# Display the filled test_df
print("Test DataFrame with predictions:")
print(filled_test_df.head())

# Check for any missing predictions
missing_predictions = filled_test_df[filled_test_df[target_column].isna()]
if len(missing_predictions) > 0:
    print(f"Number of missing predictions: {len(missing_predictions)}")
    print("Sample of rows with missing predictions:")
    print(missing_predictions.head())

# Save the filled test_df to CSV
filled_test_df.to_csv('test_predictions.csv', index=False)
print("Predictions saved to 'test_predictions.csv'")

# Create evaluation visualizations
# Plot distribution of predictions
plt.figure(figsize=(10, 6))
plt.hist(filled_test_df[target_column].dropna(), bins=30)
plt.title(f'Distribution of {target_column} Predictions')
plt.xlabel(target_column)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot predictions by kingdom group
plt.figure(figsize=(15, 8))
for kg in filled_test_df['kingdom_group'].unique():
    kg_data = filled_test_df[filled_test_df['kingdom_group'] == kg]
    plt.plot(kg_data['Date'], kg_data[target_column], label=kg)

plt.title(f'{target_column} Predictions by Kingdom Group')
plt.xlabel('Date')
plt.ylabel(target_column)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# To calculate metrics, you'll need a validation set
# Let's split the original training data
train_subset, val_subset = train_test_split(train_data, test_size=0.2, random_state=42)

# Dictionary to store models and their performance metrics
models = {}
metrics = {}

# Target variables to predict
targets = ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed', 'Wind_Direction']

# Features to use for prediction
features = ['kingdom_group', 'DayOfWeek_sin', 'DayOfWeek_cos', 'DayOfYear_sin', 'DayOfYear_cos',
            'Month_sin', 'Month_cos', 'Quarter_sin', 'Quarter_cos']

# Function to prepare data for LSTM (reshape to 3D: [samples, time steps, features])
def prepare_lstm_data(X, y, time_steps=1):
    """
    Reshape input data into 3D format required for LSTM
    X: features DataFrame
    y: target Series
    time_steps: number of time steps to look back
    """
    X_values = X.values
    y_values = y.values

    X_lstm = []
    y_lstm = []

    for i in range(len(X_values) - time_steps):
        X_lstm.append(X_values[i:i + time_steps])
        y_lstm.append(y_values[i + time_steps])

    return np.array(X_lstm), np.array(y_lstm)

# Define the LSTM model architecture
def create_lstm_model(input_shape, output_shape=1):
    """
    Create an LSTM model
    input_shape: tuple of (time_steps, features)
    """
    model = Sequential()
    model.add(LSTM(64, input_shape=input_shape, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(32, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(output_shape))

    model.compile(optimizer='adam', loss='mse')
    return model

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# Set the number of time steps to consider
time_steps = 3

# Train and evaluate a model for each target
for target in targets:
    print(f"Training and evaluating LSTM model for {target}")

    # Prepare data for LSTM
    X_train_lstm, y_train_lstm = prepare_lstm_data(
        train_subset[features], train_subset[target], time_steps
    )
    X_val_lstm, y_val_lstm = prepare_lstm_data(
        val_subset[features], val_subset[target], time_steps
    )

    # Create and train the model
    input_shape = (time_steps, len(features))
    model = create_lstm_model(input_shape)
    history = model.fit(
        X_train_lstm, y_train_lstm,
        epochs=50,
        batch_size=32,
        validation_data=(X_val_lstm, y_val_lstm),
        callbacks=[early_stopping],
        verbose=1
    )

    # Store the model
    models[target] = model

    # Make predictions on validation set
    val_predictions = model.predict(X_val_lstm)

    # Calculate metrics
    mae = mean_absolute_error(y_val_lstm, val_predictions)
    mse = mean_squared_error(y_val_lstm, val_predictions)
    rmse = sqrt(mse)
    r2 = r2_score(y_val_lstm, val_predictions)

    # Calculate MAPE (Mean Absolute Percentage Error)
    # Avoiding division by zero
    actual = y_val_lstm
    mape = np.mean(np.abs((actual - val_predictions.flatten()) / np.maximum(np.abs(actual), 1e-10))) * 100

    # Store metrics
    metrics[target] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R²': r2,
        'MAPE (%)': mape
    }

    # Plot training history
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'LSTM Training History for {target}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

# Display metrics as a DataFrame
metrics_df = pd.DataFrame(metrics)
print("\nModel Performance Metrics:")
print(metrics_df)

# Average performance across all targets
print("\nAverage Performance Across All Targets:")
avg_metrics = metrics_df.mean(axis=1)
for metric, value in avg_metrics.items():
    print(f"{metric}: {value:.4f}")

# Prepare test data for LSTM
def prepare_test_data_lstm(data, time_steps=1):
    """
    Prepare test data for LSTM prediction
    """
    X_values = data[features].values
    X_lstm = []

    # For each test sample, get the previous time_steps samples
    # Note: This assumes the test data follows the training data in chronological order
    for i in range(len(X_values) - time_steps + 1):
        X_lstm.append(X_values[i:i + time_steps])

    return np.array(X_lstm)

# Prepare test data
X_test_lstm = prepare_test_data_lstm(test_data, time_steps)

# Make predictions for test data for each target
for target in targets:
    test_predictions = models[target].predict(X_test_lstm)

    # Store predictions in test data
    # Note: Since we're using time_steps, we'll have fewer predictions than original test points
    # Adjust accordingly or use a different approach if needed
    test_data[target] = np.pad(
        test_predictions.flatten(),
        (time_steps - 1, 0),
        'constant',
        constant_values=np.nan
    )[:len(test_data)]  # Trim to original length

# If you need to inverse transform the predictions
scaled_predictions = test_data[targets].values
original_scale_predictions = scaler.inverse_transform(scaled_predictions)

for i, target in enumerate(targets):
    test_data[target] = original_scale_predictions[:, i]
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, MaxPooling1D, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

# Define the targets and features
targets = ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed', 'Wind_Direction_cos','Wind_Direction_sin']
features = ['kingdom_group', 'DayOfWeek_sin', 'DayOfWeek_cos', 'DayOfYear_sin',
            'DayOfYear_cos', 'Month_sin', 'Month_cos', 'Quarter_sin', 'Quarter_cos']

# Function to prepare time series data for CNN
def create_sequences(data, seq_length):
    """Transform the data into sequences of the specified length"""
    xs, ys = [], []
    for kingdom in data['kingdom_group'].unique():
        kingdom_data = data[data['kingdom_group'] == kingdom]

        # Sort by date for time series
        kingdom_data = kingdom_data.sort_values('Date')

        # Get feature and target values
        X = kingdom_data[features].values
        y = kingdom_data[targets].values

        # Create sequences
        for i in range(len(X) - seq_length):
            xs.append(X[i:(i + seq_length)])
            ys.append(y[i + seq_length])

    return np.array(xs), np.array(ys)

# Prepare the data
seq_length = 30  # Using 30 days of history to predict the next day

# Convert date strings to datetime objects if needed
train_df['Date'] = pd.to_datetime(train_df['Date'])
test_df['Date'] = pd.to_datetime(test_df['Date'])

# Scale the features and targets
feature_scaler = StandardScaler()
train_df[features] = feature_scaler.fit_transform(train_df[features])
test_df[features] = feature_scaler.transform(test_df[features])

target_scaler = StandardScaler()
train_df[targets] = target_scaler.fit_transform(train_df[targets])

# Create sequences for training
X_train, y_train = create_sequences(train_df, seq_length)

# Build CNN model
def build_cnn_model(input_shape, output_dim):
    model = Sequential([
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Conv1D(filters=128, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(output_dim)
    ])

    model.compile(
        optimizer='adam',
        loss='mse'
    )
    return model

# Train the model
input_shape = (seq_length, len(features))
model = build_cnn_model(input_shape, len(targets))

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1
)

# Update the predict_for_test function to ensure proper data types
def predict_for_test(model, test_df, train_df, seq_length):
    """Make predictions for test data by kingdom"""
    predictions = pd.DataFrame()

    for kingdom in test_df['kingdom_group'].unique():
        # Get the last seq_length days from training data for this kingdom
        kingdom_train = train_df[train_df['kingdom_group'] == kingdom].sort_values('Date').tail(seq_length)
        kingdom_test = test_df[test_df['kingdom_group'] == kingdom].sort_values('Date')

        # Initialize sequence with last training days - ensure it's numeric
        sequence = kingdom_train[features].values.astype(np.float32)

        # Make predictions day by day
        kingdom_preds = []
        for _, row in kingdom_test.iterrows():
            # Reshape sequence for model input
            X = np.array([sequence[-seq_length:]]).astype(np.float32)

            # Ensure X has the right shape and dtype
            if X.shape[1:] != (seq_length, len(features)):
                # If we don't have enough history, pad with zeros
                if X.shape[1] < seq_length:
                    padding = np.zeros((1, seq_length - X.shape[1], X.shape[2]), dtype=np.float32)
                    X = np.concatenate([padding, X], axis=1)

            # Make prediction
            pred = model.predict(X, verbose=0)[0]
            kingdom_preds.append(pred)

            # Update sequence with new features
            new_features = row[features].values.reshape(1, -1).astype(np.float32)
            sequence = np.vstack([sequence, new_features])

        # Create dataframe with predictions
        kingdom_df = kingdom_test.copy()

        # Only proceed if we have predictions
        if kingdom_preds:
            kingdom_pred_array = np.array(kingdom_preds)

            for i, target in enumerate(targets):
                kingdom_df[target] = kingdom_pred_array[:, i]

            predictions = pd.concat([predictions, kingdom_df])

    return predictions

# Debug function to inspect data types
def debug_data_types(df, column_list):
    print("Data Types Check:")
    for col in column_list:
        print(f"{col}: {df[col].dtype}")
        # Check for object dtypes and print sample
        if df[col].dtype == 'object':
            print(f"  Sample values: {df[col].head(3).tolist()}")
            print(f"  Contains NaN: {df[col].isna().any()}")

    # Check for any NaN values in the dataframe
    print(f"DataFrame contains NaN: {df.isna().any().any()}")
    if df.isna().any().any():
        print("Columns with NaN:")
        print(df.isna().sum()[df.isna().sum() > 0])

# Use this before training to check data types
print("Checking training data types:")
debug_data_types(train_df, features + targets)

# Fix any object or mixed dtypes in the dataframes
def fix_datatypes(df, columns):
    for col in columns:
        if df[col].dtype == 'object':
            try:
                df[col] = df[col].astype(float)
            except ValueError:
                print(f"Could not convert {col} to float. Sample values: {df[col].head(3).tolist()}")
    return df

# Apply fixes to both dataframes
train_df = fix_datatypes(train_df, features + targets)
test_df = fix_datatypes(test_df, features)

# Now try the prediction again
try:
    predictions_df = predict_for_test(model, test_df, train_df, seq_length)

    # Inverse transform the predictions to original scale
    predictions_array = predictions_df[targets].values
    original_scale_predictions = target_scaler.inverse_transform(predictions_array)

    for i, target in enumerate(targets):
        predictions_df[target] = original_scale_predictions[:, i]

    print("Predictions completed successfully!")
except Exception as e:
    print(f"Error during prediction: {type(e).__name__}")
    print(e)

    # Additional debugging information
    print("\nSequence shape information:")
    # Get a sample kingdom to debug
    sample_kingdom = test_df['kingdom_group'].unique()[0]
    kingdom_train = train_df[train_df['kingdom_group'] == sample_kingdom].sort_values('Date').tail(seq_length)
    print(f"Sample sequence shape: {kingdom_train[features].values.shape}")
    print(f"Expected input shape for model: {(seq_length, len(features))}")

# Evaluate the model on a validation set
val_size = int(0.1 * len(train_df))
val_df = train_df.tail(val_size)
train_df_reduced = train_df.iloc[:-val_size]

X_train_final, y_train_final = create_sequences(train_df_reduced, seq_length)
val_predictions = predict_for_test(model, val_df, train_df_reduced, seq_length)

# Calculate metrics
metrics = {}
for i, target in enumerate(targets):
    actual = target_scaler.inverse_transform(val_df[targets].values)[:, i]
    predicted = target_scaler.inverse_transform(val_predictions[targets].values)[:, i]

    mae = mean_absolute_error(actual, predicted)
    mse = mean_squared_error(actual, predicted)
    rmse = sqrt(mse)
    r2 = r2_score(actual, predicted)

    # Calculate MAPE (Mean Absolute Percentage Error)
    mape = np.mean(np.abs((actual - predicted) / np.maximum(np.abs(actual), 1e-10))) * 100

    metrics[target] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R²': r2,
        'MAPE (%)': mape
    }

# Display metrics
metrics_df = pd.DataFrame(metrics)
print("\nModel Performance Metrics:")
print(metrics_df)

# Average performance across all targets
print("\nAverage Performance Across All Targets:")
avg_metrics = metrics_df.mean(axis=1)
for metric, value in avg_metrics.items():
    print(f"{metric}: {value:.4f}")

# Save predictions to test_df
for target in targets:
    test_df[target] = predictions_df[target]

# You can now use test_df with the predicted values
print("Predictions completed and added to test_df")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, MaxPooling1D, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

# Define the targets and features
targets = ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed', 'Wind_Direction_cos','Wind_Direction_sin']
features = ['DayOfWeek_sin', 'DayOfWeek_cos', 'DayOfYear_sin',
            'DayOfYear_cos', 'Month_sin', 'Month_cos', 'Quarter_sin', 'Quarter_cos']

# Function to prepare time series data for CNN
def create_sequences(data, seq_length):
    """Transform the data into sequences of the specified length"""
    xs, ys = [], []

    # Sort by date for time series
    data = data.sort_values('Date')

    # Get feature and target values
    X = data[features].values
    y = data[targets].values

    # Create sequences
    for i in range(len(X) - seq_length):
        xs.append(X[i:(i + seq_length)])
        ys.append(y[i + seq_length])

    return np.array(xs), np.array(ys)

# Build CNN model
def build_cnn_model(input_shape, output_dim):
    model = Sequential([
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Conv1D(filters=128, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(output_dim)
    ])

    model.compile(
        optimizer='adam',
        loss='mse'
    )
    return model

# Function to predict for a kingdom (including handling early dates)
def predict_for_kingdom_complete(model, kingdom_test, kingdom_train, scaler_X, scaler_y, seq_length):
    """Make predictions for a specific kingdom, ensuring all test dates are predicted"""
    # Create a copy of test data that will contain our predictions
    test_predictions = kingdom_test.copy().sort_values('Date')

    # Add target columns with NaN values if they don't exist
    for target in targets:
        if target not in test_predictions.columns:
            test_predictions[target] = np.nan

    # We need to handle two scenarios:
    # 1. For dates where we have enough history, use normal prediction
    # 2. For early dates where we don't have enough history, we need a different approach

    # For predictions where we have enough history
    # Create a copy of kingdom_train with only the features and Date
    history_data = kingdom_train[features + ['Date']].copy()

    # Make sure history data is sorted by date
    history_data = history_data.sort_values('Date')

    # For each test row, predict using the last seq_length days
    for i in range(len(test_predictions)):
        # Find the date of this test row
        test_date = test_predictions.iloc[i]['Date']

        # Get data before this date that we can use for prediction
        history_before_test = history_data[history_data['Date'] < test_date]

        # If we have enough history, make a normal prediction
        if len(history_before_test) >= seq_length:
            # Get the last seq_length days as our sequence
            sequence = history_before_test.tail(seq_length)[features].values
            sequence = np.array([sequence])  # Reshape for the model

            # Make prediction
            pred_scaled = model.predict(sequence, verbose=0)[0]
            # Inverse transform
            pred = scaler_y.inverse_transform(pred_scaled.reshape(1, -1))[0]

            # Store prediction in our results DataFrame
            for j, target in enumerate(targets):
                test_predictions.iloc[i, test_predictions.columns.get_loc(target)] = pred[j]
        else:
            # For early dates without enough history, use a simple baseline approach

            # Get the average of available training data for this date's month
            test_month = test_date.month
            month_data = kingdom_train[pd.to_datetime(kingdom_train['Date']).dt.month == test_month]

            if len(month_data) > 0:
                # Use monthly average from training data
                for j, target in enumerate(targets):
                    if target in month_data.columns:  # Make sure the target exists in month_data
                        test_predictions.iloc[i, test_predictions.columns.get_loc(target)] = month_data[target].mean()
            else:
                # Fallback: use overall average from training data
                for j, target in enumerate(targets):
                    if target in kingdom_train.columns:  # Make sure the target exists in kingdom_train
                        test_predictions.iloc[i, test_predictions.columns.get_loc(target)] = kingdom_train[target].mean()

        # Add this test row's features to our history for future predictions
        current_features = test_predictions.iloc[[i]][features + ['Date']]
        history_data = pd.concat([history_data, current_features])
        history_data = history_data.sort_values('Date')

    return test_predictions

# Main function to train separate models and make predictions
def train_and_predict_by_kingdom(train_df, test_df, seq_length=30, val_split=0.2, epochs=100):
    # Convert date strings to datetime objects if needed
    train_df['Date'] = pd.to_datetime(train_df['Date'])
    test_df['Date'] = pd.to_datetime(test_df['Date'])

    # Check if test_df has the target columns, if not, add them with NaN values
    for target in targets:
        if target not in test_df.columns:
            test_df[target] = np.nan

    all_predictions = pd.DataFrame()
    all_metrics = {}

    # Train separate models for each kingdom
    kingdom_groups = train_df['kingdom_group'].unique()

    print(f"Training separate models for {len(kingdom_groups)} kingdom groups...")

    for kingdom in kingdom_groups:
        print(f"\nProcessing kingdom group: {kingdom}")

        # Get data for this kingdom
        kingdom_train = train_df[train_df['kingdom_group'] == kingdom].copy()
        kingdom_test = test_df[test_df['kingdom_group'] == kingdom].copy()

        if len(kingdom_train) <= seq_length:
            print(f"Insufficient training data for kingdom {kingdom}. Using baseline predictions.")
            # For small kingdoms, use baseline predictions (mean values)
            for target in targets:
                if target in kingdom_train.columns:
                    kingdom_test[target] = kingdom_train[target].mean() if len(kingdom_train) > 0 else 0
                else:
                    # If targets don't exist in train data either, use zeros
                    kingdom_test[target] = 0

            kingdom_test['kingdom_group'] = kingdom
            all_predictions = pd.concat([all_predictions, kingdom_test])
            continue

        # Split for validation
        val_size = max(1, int(val_split * len(kingdom_train)))
        kingdom_val = kingdom_train.tail(val_size).copy()
        kingdom_train_final = kingdom_train.iloc[:-val_size].copy()

        # Store original values before scaling
        kingdom_train_orig = kingdom_train_final.copy()

        # Scale features and targets for this kingdom
        feature_scaler = StandardScaler()
        target_scaler = StandardScaler()

        # Fit scalers on training data
        kingdom_train_final[features] = feature_scaler.fit_transform(kingdom_train_final[features])
        kingdom_train_final[targets] = target_scaler.fit_transform(kingdom_train_final[targets])

        # Transform validation data
        kingdom_val[features] = feature_scaler.transform(kingdom_val[features])
        kingdom_val[targets] = target_scaler.transform(kingdom_val[targets])

        # Create sequences for training
        X_train, y_train = create_sequences(kingdom_train_final, seq_length)

        if len(X_train) == 0:
            print(f"Not enough sequence data for kingdom {kingdom} after creating sequences. Using baseline predictions.")
            # Use baseline predictions
            for target in targets:
                if target in kingdom_train.columns:
                    kingdom_test[target] = kingdom_train[target].mean()
                else:
                    kingdom_test[target] = 0

            kingdom_test['kingdom_group'] = kingdom
            all_predictions = pd.concat([all_predictions, kingdom_test])
            continue

        # Initialize and train model
        input_shape = (seq_length, len(features))
        kingdom_model = build_cnn_model(input_shape, len(targets))

        # Setup early stopping
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )

        # Create validation sequences
        X_val, y_val = create_sequences(kingdom_val, seq_length)

        # Train model if we have validation data
        if len(X_val) > 0:
            history = kingdom_model.fit(
                X_train, y_train,
                epochs=epochs,
                batch_size=min(32, len(X_train)),
                validation_data=(X_val, y_val),
                callbacks=[early_stopping],
                verbose=1
            )

            # Calculate validation metrics if we have validation data
            # Create a copy of kingdom_val for prediction to avoid modifying the original
            val_for_pred = kingdom_val.copy()

            # Make sure val_for_pred has all target columns (filled with NaN if needed)
            for target in targets:
                if target not in val_for_pred.columns:
                    val_for_pred[target] = np.nan

            val_pred_df = predict_for_kingdom_complete(
                kingdom_model, val_for_pred, kingdom_train_final,
                feature_scaler, target_scaler, seq_length
            )

            # Calculate metrics
            kingdom_metrics = {}
            for target in targets:
                if target in kingdom_val.columns and target in val_pred_df.columns:
                    # Only use rows that have both actual and predicted values
                    valid_indices = ~(kingdom_val[target].isna() | val_pred_df[target].isna())
                    actual = kingdom_val.loc[valid_indices, target].values
                    predicted = val_pred_df.loc[valid_indices, target].values

                    if len(actual) > 0 and len(predicted) > 0:
                        mae = mean_absolute_error(actual, predicted)
                        mse = mean_squared_error(actual, predicted)
                        rmse = sqrt(mse)
                        r2 = r2_score(actual, predicted) if len(actual) > 1 else np.nan

                        # Calculate MAPE
                        mape = np.mean(np.abs((actual - predicted) / np.maximum(np.abs(actual), 1e-10))) * 100

                        kingdom_metrics[target] = {
                            'MAE': mae,
                            'MSE': mse,
                            'RMSE': rmse,
                            'R²': r2,
                            'MAPE (%)': mape
                        }

            # Store metrics
            all_metrics[kingdom] = kingdom_metrics
            print(f"Validation metrics for kingdom {kingdom}:")
            for target, metrics in kingdom_metrics.items():
                print(f"  {target}: MAE={metrics['MAE']:.4f}, RMSE={metrics['RMSE']:.4f}, R²={metrics['R²']:.4f}")
        else:
            # If no validation data, just train on all available data
            history = kingdom_model.fit(
                X_train, y_train,
                epochs=epochs,
                batch_size=min(32, len(X_train)),
                verbose=1
            )

        # Make predictions for test data
        if not kingdom_test.empty:
            # Make predictions using our complete prediction function
            kingdom_pred_df = predict_for_kingdom_complete(
                kingdom_model, kingdom_test, kingdom_train_orig,
                feature_scaler, target_scaler, seq_length
            )

            # Add kingdom group back to predictions
            kingdom_pred_df['kingdom_group'] = kingdom

            # Add to all predictions
            all_predictions = pd.concat([all_predictions, kingdom_pred_df])

    # Compute average metrics across kingdoms
    if all_metrics:
        print("\nAverage Metrics Across All Kingdom Groups:")
        avg_metrics = {}
        for metric_type in ['MAE', 'MSE', 'RMSE', 'R²', 'MAPE (%)']:
            avg_metrics[metric_type] = {}
            for target in targets:
                values = [metrics[target][metric_type] for kingdom, metrics in all_metrics.items()
                         if target in metrics and np.isfinite(metrics[target][metric_type])]
                avg_metrics[metric_type][target] = np.mean(values) if values else np.nan

            print(f"{metric_type}:")
            for target, value in avg_metrics[metric_type].items():
                print(f"  {target}: {value:.4f}")

    # Verify we have predictions for all test rows
    print(f"\nNumber of predictions: {len(all_predictions)}")
    print(f"Number of test rows: {len(test_df)}")

    if len(all_predictions) != len(test_df):
        print("WARNING: Not all test rows have predictions!")

        # Count predictions per kingdom
        kingdom_counts = all_predictions['kingdom_group'].value_counts()
        test_kingdom_counts = test_df['kingdom_group'].value_counts()

        print("\nPredictions per kingdom vs Test data per kingdom:")
        for kingdom in test_df['kingdom_group'].unique():
            pred_count = kingdom_counts.get(kingdom, 0)
            test_count = test_kingdom_counts.get(kingdom, 0)
            print(f"Kingdom {kingdom}: {pred_count} predictions / {test_count} test rows")

    return all_predictions, all_metrics

# Run the kingdom-specific training and prediction
seq_length = 30  # Using 30 days of history to predict the next day
predictions_df, metrics = train_and_predict_by_kingdom(train_df, test_df, seq_length=seq_length)

# Verify we have the expected number of rows
expected_rows = len(test_df)
actual_rows = len(predictions_df)
print(f"\nExpected rows: {expected_rows}, Actual rows: {actual_rows}")

# Ensure we have all required columns in the output
for target in targets:
    if target not in predictions_df.columns:
        print(f"WARNING: Target column {target} is missing from predictions!")

# Save predictions to CSV
predictions_df.to_csv('kingdom_specific_predictions.csv', index=False)
print("\nPredictions saved to 'kingdom_specific_predictions.csv'")

# Optional: Detailed metrics per kingdom
print("\nDetailed Metrics by Kingdom:")
for kingdom, kingdom_metrics in metrics.items():
    print(f"\nKingdom: {kingdom}")
    for target, target_metrics in kingdom_metrics.items():
        print(f"  {target}:")
        for metric_name, metric_value in target_metrics.items():
            print(f"    {metric_name}: {metric_value:.4f}")
```
